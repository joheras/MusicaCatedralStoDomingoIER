```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('portada.tif')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1,1,figsize=(16,16))
ax.imshow(img,cmap=plt.cm.binary)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
plt.show()
```

```python
ins = img[2550:4550,720:3960]
fig, ax = plt.subplots(1,1,figsize=(16,16))
ax.imshow(ins,cmap=plt.cm.binary)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
plt.show()
```

```python
tit = img[550:2550,730:3970]
fig, ax = plt.subplots(1,1,figsize=(16,16))
ax.imshow(tit,cmap=plt.cm.binary)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
plt.show()
```

```python
port = np.concatenate([tit,ins],axis=1)
fig, ax = plt.subplots(1,1,figsize=(16,16))
ax.imshow(port,cmap=plt.cm.binary)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
plt.show()
```

```python
port = img = cv2.cvtColor(port,cv2.COLOR_RGB2BGR)
cv2.imwrite('port.png',port)
```
